# Exploration and Interpretation

Throughout this project, we use book id, chapter number, sentence number, and token number as the OHCO. We used TF-IDF (Term Frequency-Inverse Document Frequency) in order to measure the importance of words and terms in the corpus. We also used PCA, LDA, Word Embeddings, and Sentiment Analysis. 

### Principal Component Analysis (PCA)

For this project, we wanted to understand the main themes in the Harry Potter books as well as the similarity between books. One of the ways we explored this question is through the use of PCA. We grouped the main ideas and words together in order to learn from our analysis. 

We used PCA on the TFIDF matrix to reduce dimensionality of the data and find the most important patterns and relationships between variables. Here, our variables are the words in the Harry Potter books and our goal is indentifying the most significant themes and motifs in these books. 

The code normalizes the TFIDF matrix, then it calculates the covariance matrix, and then decomposes this matrix into eigenvalues and eigenvectors. This is done so that we can identify the most important patterns in our data (given by the eigenvectors) and the variance in the data (explained by the eigenvalues). 


### Hierarchical Cluster Analysis (HCA)

We want to see the relationship between the Harry Potter books. A good way to do this is using hierarchical cluster diagrams. HCA (Hierarchical cluster analysis) allows us to learn about the relationship between terms as well as identify groups of terms or clusters that are similar in either their usage or meaning. 

We implement HCA using cosine similarity as the similarity metric. Cosine similarity ranges from -1 to 1, where 1 means high similarity and -1 means low similarity. We use a dendogram to show the similarity between the different books. We also explore other similarity metrics and display this information in a heatmap. 


```python
# for visuals
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# HCA using Cosine similarity metric and Ward as the linkage method

# HCA_Cosine_Ward_Dendo.png

# Load the image from file
image = mpimg.imread('Visuals/HCA_Cosine_Ward_Dendo.png')

# size
plt.figure(figsize=(12,12))

# Display the image
plt.imshow(image)
plt.show()
```

In the above dendogram, we used Cosine as the similarity metric and Ward as the linkage method. Here we see several interesting results:

1. Goblet of fire (book 4), Prisoner of Azkaban (book 3), Deathly Hallows (book 7), and Half-Blood Prince (book 6) are all seen to be very closely related 
2. Specifically, Goblet of fire (book 4) and Prisoner of Azkaban (book 3) are the most closely linked. 
3. Interestingly, Order of the Phoenix (book 5), Sorcerer's Stone (book 1), and Chamber of Secrets (book 2) are not very closely linked to each other, relative to the other four books

So, we see books 3, 4, 6, and 7 are in one main cluster, very closely linked, and books 1, 2, and 5 are in another main cluster, not as closely linked. These clusters show somewhat of a division between earlier and later books, where later books are much more closely linked with each other.

```python
# HCA heatmap with multiple different similarity metrics

# HCA_Heatmap.png

# Load the image from file
image = mpimg.imread('Visuals/HCA_Heatmap.png')

# size
plt.figure(figsize=(12,12))

# Display the image
plt.imshow(image)
plt.show()
```

In the above heatmap, we compare the similarity of the different books with each other by several different similarity metrics: 

1. Cityblock
2. Euclidean
3. Cosine
4. Jaccard
5. Dice
6. JS (Jensen-Shannon divergence)

This heatmap uses similarity as color and is sorted by cosine similarity. We are able to see that the darker cells correspond with higher similarity and the lighter cells correspond with lower similarity. 

We see the following pairs of metrics providing the most similar results: 
1. Cityblock and Euclidean
2. Jaccard and Dice
3. Cosine and Jensen-Shannon divergence

Despite the differences between methods, overall the heatmap does yield results that support the idea of books 3, 4, 6, 7 being relatively similar, and books 1, 2, 5 being relatively not similar.

### Sentiment Analysis

In this section, we want to explore the sentiment of the seven books and how it changes over time. This will help us learn more about how the books change over time, as characters and the plot change and develop. We used sentiment analysis grouped by chapter, throughout the seven books. Then, we look at trust, fear, and joy, throughout time. This is because we want to get an understanding of how these qualities change over time and how they relate to each other and the overall sentiment. 

We will display a visual to showcase this below. 

```python
# Sentiment visual

# Sentiment_over_time.png

# Load the image from file
image = mpimg.imread('Visuals/Sentiment_over_time.png')

# size
plt.figure(figsize=(30,30))

# Display the image
plt.imshow(image)
plt.show()
```

As we see in the above, there are dramatic changes in sentiment over time. We see many peaks and troughs, showing that the series has a sequence of positive and negative shifts of sentiment. This makes sense, as it would build tension and cause the reader to be more intrigued.

The books must maintain their reader's interest, and typically we see negative emotions caputuring more attention than positive ones. Furthermore, it makes sense that negative sentiment would keep the reader more captivated, especially in a series that is designed such that there is an evil villian who causes chaos and who the heros must eventually defeat.

Furthermore, we see at the troughs of sentiment, a very large spike in fear and drop in trust and joy, which is to be expected. Overall, there is a strong postive correlation between trust and joy. We also see that fear seems inversely correlated with trust and joy.

Originally, I was hoping to see some sort of trends persisting throughout time. However, it seems like throughout the story and plot of the book, there is not much consistency, with sentiment appearing to randomly ocillate over time. 

### Word Embeddings

Here, we use word embeddings to see the relationship between different words in the Harry Potter series. This will help us discover the similarity between the different words thus expanding on our previous analysis of the similarity between the books in the series.
